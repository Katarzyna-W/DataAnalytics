# Modelling GDP growth

Data on average growth rates over 1960â€“1995 for 65 countries, along with variables that are potentially related to growth.
The task is to create a model to represent economic growth, assuming an average growth rate of ~2%, and then examine whether the data fits the model created.

Author: Kamil Gniadek
Date: 09.06.2021

```python
from cmdstanpy import CmdStanModel

import arviz as az
import numpy as np
import scipy.stats as stats

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.context('seaborn-white')
mpl.rcParams['figure.dpi']= 200
```

```python
# Adding some extra colors for ploting
LIGHT="#FFFCDC"
LIGHT_HIGHLIGHT="#FEF590"
MID="#FDED2A"
MID_HIGHLIGHT="#f0dc05"
DARK="#EECA02"
DARK_HIGHLIGHT="#BB9700"
GREEN="#00FF00"
LIGHT_GREY="#DDDDDD"
```

### Loading data

```python
data = pd.read_csv('input\GrowthSW.csv', sep=',', header=0)
data.head()
```

## Normal model - no predictors

### Prior predictive checks

Assuming that the average growth is around 2%, we create a model that will take this value into account and be tolerant of the subsequent impact of the data.

```python
with open('growth_1_ppc.stan', 'r') as file:
    print(file.read())
```

```python
growth_1_ppc_model=CmdStanModel(stan_file='growth_1_ppc.stan')
```

```python
R = 1000
sim=growth_1_ppc_model.sample(iter_sampling=R, iter_warmup=0,chains=1,fixed_param=True,seed=29042020)
```

```python
mu_sim=sim.stan_variable('mu')
sigma_sim=sim.stan_variable('sigma')
growth_prior_sim=sim.stan_variable('growth_prior')
```

```python
fig = plt.figure(figsize=(7, 4))

gs = fig.add_gridspec(2,2)
ax1 = fig.add_subplot(gs[1, :])
ax1.hist(growth_prior_sim,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax1.set_title('Simulated growth')

ax2 = fig.add_subplot(gs[0, 0])
ax3 = fig.add_subplot(gs[0, 1])
ax2.set_yticks([])
ax2.hist(mu_sim,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
x=np.linspace(-10,10)
y=stats.norm.pdf(x,loc=2,scale=3)
ax2.plot(x,y,'black')
ax2.set_title('$\mu$')
ax3.set_yticks([])
ax2.set_yticks([])
ax3.hist(sigma_sim,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
x=np.linspace(0,20)
y=stats.expon.pdf(x,scale=3.33)
ax3.plot(x,y,color='black')
ax3.set_title('$\sigma$')

fig.tight_layout()
plt.show()
```

### Posterior and inference

```python
with open('growth_1_fit.stan', 'r') as file:
    print(file.read())
```

```python
growth_1_fit_model=CmdStanModel(stan_file='growth_1_fit.stan')
```

```python
growth_1_fit=growth_1_fit_model.sample(data=dict(N=len(data),growth_data=data.growth.values),seed=28052020)
```

```python
az.summary(growth_1_fit,var_names=['mu','sigma'],round_to=2,kind='stats')
```

```python
mu_fit=growth_1_fit.stan_variable('mu')
sigma_fit=growth_1_fit.stan_variable('sigma')
growth_pred=growth_1_fit.stan_variable('growth_posterior')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(7, 4),sharex=True,sharey=True)
ax=axes[0]
ax.hist(data.growth,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax.set_title('Observed growth')
ax.set_yticks(())
ax2=axes[1]
ax2.hist(growth_pred,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax2.hist(data.growth,bins=20,histtype='step',color='black',density=True)

ax2.set_title('Predicted growth')
ax2.set_yticks(())
fig.tight_layout()

plt.show()
```

## Improving model - adding a education predictor
Previous results have satisfactorily reproduced the actual data, but we would like the model to be even more accurate. In this section, I will introduce a predictor in the form of one of the resulting data: the average number of years of schooling of adult residents in that country in 1960.


### Centering of variable
It is used for easier interpretation of parameters.

```python
data['c_education']=data.education-data.education.mean()
data.head()
```

### Prior predictive checks

```python
with open('growth_2_education_ppc.stan', 'r') as file:
    print(file.read())
```

```python
growth_2_education_ppc_model=CmdStanModel(stan_file='growth_2_education_ppc.stan')
```

```python
R = 1000
data_sim=dict(N=len(data),education_length=data.c_education.values)
sim=growth_2_education_ppc_model.sample(data=data_sim, 
                     iter_sampling=R, 
                     iter_warmup=0, 
                     chains=1, 
                     fixed_param=True,
                     seed=29042020)
```

```python
alpha_sim=sim.stan_variable('alpha')
beta_sim=sim.stan_variable('beta')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 4))
for i in range(100):
    axes.plot(data.education,alpha_sim[i]+beta_sim[i]*data.c_education,color=MID,alpha=0.5,linewidth=0.5)
axes.set_xlabel('education length [years]')
axes.set_ylabel('growth [%]')
axes.set_title(r'Prior for $\beta$')
plt.show()
```

### Posterior inference

```python
with open('growth_2_education_fit.stan', 'r') as file:
    print(file.read())
```

```python
growth_2_education_fit_model=CmdStanModel(stan_file='growth_2_education_fit.stan')
```

```python
data_fit=dict(N=len(data),
              education_length=data.c_education.values,
              growth_data=data.growth.values)

growth_2_education_fit=growth_2_education_fit_model.sample(data=data_fit,seed=28052020)
```

```python
alpha_fit=growth_2_education_fit.stan_variable('alpha')
beta_fit=growth_2_education_fit.stan_variable('beta')
mu_fit=growth_2_education_fit.stan_variable('mu')
growth_pred_education=growth_2_education_fit.stan_variable('growth_posterior')
```

```python
az.summary(growth_2_education_fit,var_names=['alpha','beta','sigma'],round_to=2,kind='stats')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 4))
for i in range(100):
    axes.plot(data.education,
              alpha_fit[i]+beta_fit[i]*data.c_education,
              color=MID,
              alpha=0.5,linewidth=0.5)
axes.scatter(data.education,data.growth,color='black',alpha=0.1,s=10)
axes.set_xlabel('education length [years]')
axes.set_ylabel('growth [%]')
axes.set_title(r'Linear corelation of growth and education length')

fig.tight_layout()
plt.show()
```

### Posterior predictive check

```python
model_education_id = az.from_cmdstanpy(posterior=growth_2_education_fit,
                           log_likelihood='log_lik',
                           posterior_predictive='growth_posterior',
                           observed_data={'growth_data':data['growth']} )
```

```python
ax=az.plot_ppc(data = model_education_id, alpha=0.1,color=mid_highlight,figsize=(7, 4),data_pairs={'growth_data':'growth_posterior'},num_pp_samples=100,observed=False,random_seed=123)
ax.set_xlim((-10,10))
ax.hist(data.growth,bins=np.linspace(-45,45,42),histtype='step',edgecolor='black',density=True,label='Observed data')
ax.set_title(r'Posterior predictive check - education predictor')
ax.legend()
plt.show()
```

### Compare models without predictor and with education length as predictor

```python
fig, axes = plt.subplots(1, 3, figsize=(7, 4),sharex=True,sharey=True)

ax=axes[0]
ax.hist(data.growth,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax.set_title('Observed growth')
ax.set_yticks(())

ax2=axes[1]
ax2.hist(growth_pred,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax2.hist(data.growth,bins=20,histtype='step',color='black',density=True)
ax2.set_title('Predicted growth')
ax2.set_yticks(())

ax3=axes[2]
color_array = [MID]*65
ax3.hist(growth_pred_education,bins=20,color=color_array,edgecolor=DARK_HIGHLIGHT,density=True)
ax3.hist(data.growth,bins=20,histtype='step',color='black',density=True)
ax3.set_title('Predicted growth - education')
ax3.set_yticks(())

fig.tight_layout()

plt.show()
```

## Improving model - adding a impact of tradeshare predictor
The model incorporating the length of education allowed for an even better fit of the model to the actual data. 
In this section, I would like to examine the effect of average share of trade (measured as the sum of exports (X) plus imports (M), divided by GDP) in the economy on overall economic growth. 

### Centering of variable
It is used for easier interpretation of parameters.

```python
data['c_tradeshare']=data.tradeshare-data.tradeshare.mean()
data.head()
```

### Prior predictive checks

```python
with open('growth_2_tradeshare_ppc.stan', 'r') as file:
    print(file.read())
```

```python
growth_2_tradeshare_ppc_model=CmdStanModel(stan_file='growth_2_tradeshare_ppc.stan')
```

```python
R = 1000
data_sim=dict(N=len(data),tradeshare=data.c_tradeshare.values)
sim=growth_2_tradeshare_ppc_model.sample(data=data_sim, 
                     iter_sampling=R, 
                     iter_warmup=0, 
                     chains=1, 
                     fixed_param=True,
                     seed=29042020)
```

```python
alpha_sim=sim.stan_variable('alpha')
beta_sim=sim.stan_variable('beta')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 4))
for i in range(100):
    axes.plot(data.tradeshare,alpha_sim[i]+beta_sim[i]*data.c_tradeshare,color=MID,alpha=0.5,linewidth=0.5)
axes.set_xlabel('tradeshare')
axes.set_ylabel('growth [%]')
axes.set_title(r'Prior for $\beta$')
plt.show()
```

### Posterior inference

```python
with open('growth_2_tradeshare_fit.stan', 'r') as file:
    print(file.read())
```

```python
growth_2_tradeshare_fit_model=CmdStanModel(stan_file='growth_2_tradeshare_fit.stan')
```

```python
data_fit=dict(N=len(data),
              tradeshare=data.c_tradeshare.values,
              growth_data=data.growth.values)

growth_2_tradeshare_fit=growth_2_tradeshare_fit_model.sample(data=data_fit,seed=28052020)
```

```python
alpha_fit=growth_2_tradeshare_fit.stan_variable('alpha')
beta_fit=growth_2_tradeshare_fit.stan_variable('beta')
mu_fit=growth_2_tradeshare_fit.stan_variable('mu')
growth_pred_tradeshare=growth_2_tradeshare_fit.stan_variable('growth_posterior')
```

```python
az.summary(growth_2_tradeshare_fit,var_names=['alpha','beta','sigma'],round_to=2,kind='stats')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 4))
for i in range(100):
    axes.plot(data.tradeshare,
              alpha_fit[i]+beta_fit[i]*data.c_tradeshare,
              color=MID,
              alpha=0.5,linewidth=0.5)
axes.scatter(data.tradeshare,data.growth,color='black',alpha=0.1,s=10)
axes.set_xlabel('tradeshare')
axes.set_ylabel('growth [%]')
axes.set_title(r'Linear corelation of growth and tradeshare')

fig.tight_layout()
plt.show()
```

### Posterior predictive check

```python
model_tradeshare_id = az.from_cmdstanpy(posterior=growth_2_tradeshare_fit,
                           log_likelihood='log_lik',
                           posterior_predictive='growth_posterior',
                           observed_data={'growth_data':data['growth']} )
```

```python
ax=az.plot_ppc(data = model_tradeshare_id, alpha=0.1,color=mid_highlight,figsize=(7, 4),data_pairs={'growth_data':'growth_posterior'},num_pp_samples=100,observed=False,random_seed=123)
ax.set_xlim((-10,10))
ax.hist(data.growth,bins=np.linspace(-45,45,42),histtype='step',edgecolor='black',density=True,label='Observed data')
ax.set_title(r'Posterior predictive check - tradeshare predictor')
ax.legend()
plt.show()
```

### Compare models without predictor and with 2 different predictors: education_length and tradeshare

```python
fig, axes = plt.subplots(1, 4, figsize=(11, 4),sharex=True,sharey=True)

ax=axes[0]
ax.hist(data.growth,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax.set_title('Observed growth')
ax.set_yticks(())

ax2=axes[1]
ax2.hist(growth_pred,bins=20,color=DARK,edgecolor=DARK_HIGHLIGHT,density=True)
ax2.hist(data.growth,bins=20,histtype='step',color='black',density=True)
ax2.set_title('Predicted growth')
ax2.set_yticks(())

ax3=axes[2]
color_array = [MID]*65
ax3.hist(growth_pred_education,bins=20,color=color_array,edgecolor=DARK_HIGHLIGHT,density=True)
ax3.hist(data.growth,bins=20,histtype='step',color='black',density=True)
ax3.set_title('Predicted growth - education')
ax3.set_yticks(())

ax4=axes[3]
color_array = [MID]*65
ax4.hist(growth_pred_tradeshare,bins=20,color=color_array,edgecolor=DARK_HIGHLIGHT,density=True)
ax4.hist(data.growth,bins=20,histtype='step',color='black',density=True)
ax4.set_title('Predicted growth - tradeshare')
ax4.set_yticks(())

fig.tight_layout()

plt.show()
```

#### Summary:
Tradeshare predictor allowed the model to fit better near its mean value, while it had a bad effect on the tails of the distribution.

# Prior predictive check

```python
normal_model1 = CmdStanModel(stan_file = 'normal1.stan')
R = 1000
sim = normal_model1.sample(iter_sampling = R, 
                             iter_warmup = 0,
                             chains = 1,
                             fixed_param = True,
                             seed = 11062021)
mu_sim = sim.stan_variable('mu')
sigma_sim = sim.stan_variable('sigma')
prior = sim.stan_variable('prior')
```

```python
print(mu_sim)
print(sigma_sim)
print(prior)
```

```python
california_larceny = california.filter(['Year', 'Data.Rates.Property.Larceny'], axis = 1)
california_larceny.hist(column = 'Data.Rates.Property.Larceny', bins = 30)
#print(california_larceny)
```

```python
mean_larceny = california_larceny["Data.Rates.Property.Larceny"].mean()
print(mean_larceny)
```

```python
std_larceny = california_larceny["Data.Rates.Property.Larceny"].std()
print(std_larceny)
```

```python
fig = plt.figure(figsize = (7, 4))

gs = fig.add_gridspec(2, 2)
ax1 = fig.add_subplot(gs[1, :])
ax1.hist(prior, bins = 20, density = True)
ax1.set_title('Simulated larceny rate')

ax2 = fig.add_subplot(gs[0, 0])
ax2.hist(mu_sim, bins = 20, density = True)
x = np.linspace(0, 5599)
y = stats.norm.pdf(x, loc = 2568, scale = 878)
ax2.plot(x, y, 'black')
ax2.set_title('$\mu$')
ax2.set_yticks([])

ax3 = fig.add_subplot(gs[0, 1])
ax3.hist(sigma_sim, bins = 20, density = True)
x = np.linspace(0, 0.01)
y = stats.expon.pdf(x, scale = 878)
ax3.plot(x, y, color = 'black')
ax3.set_title('$\sigma$')
ax3.set_yticks([])

fig.tight_layout()
plt.show()
```
